```python
import numpy as np
import matplotlib.pyplot as plt
def random_walk(num_steps, x_start=0):
    a_steps = np.random.randint(2, size=num_steps)
    b_steps = a_steps * 2 - 1
    return x_start + np.cumsum(b_steps)

k = 150
final_pos = np.zeros(2*k)

n=1000

for i in range(n):
    rw = random_walk(n, k)
    for i in rw:
        final_pos[i]+=1

prob = final_pos/(n*n)        

plt.plot(prob)
print(sum(prob))
```

```python
def fact(n):
    if n==1 or n==0:
        return 1
    else:
        return n*fact(n-1)
def ncmcalc(n,m):
    return (fact(n))/(fact(m)*fact(n-m))    
```
