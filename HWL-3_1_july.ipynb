### Random Walk
1. Write a functin that will output a trajectory o unbiased random walk: start at x=0, randomly increment/decrement by 1. Total of N steps: function random_walk(num_steps,x_start=0)
    - plot a few trajectories
2. Multiple realisations require statistics.
    - get statistics and end points of an unbaised randomwalk (i.e., x_n vs Prob(x_n))
    - get <x_n> and <x_n^2>; plot both, i.e., <x_n> vs <x_n^2> vs n
3. Caluculate p

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def randwalk(num_steps,x_start=0):
    steps=2*np.random.randint(2,size=num_steps)-1
    return x_start+np.cumsum(steps)
```

```python
plt.plot(randwalk(100,10))
```

```python
for i in range(100):
    plt.plot(randwalk(100,10))
```

```python
randwalk(10,0)
```

```python
n=100
xi = np.zeros(n)
xi_sq = np.zeros(n)
for i in range(n):
    rw1=randwalk(n)
    xi += rw1
    xi_sq += (rw1*rw1)
    plt.plot(rw1)
axi = xi/n
axi_sq=xi_sq/n
```
