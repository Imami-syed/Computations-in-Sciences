```python
# function to caluculate factorial
'''Returns the value of n!
    Call: factorial(n)
'''
def factorial(n):
    f=1
    for i in range(2,n+1):
        f*=i
    return f 
```

```python
factorial(5)
```

```python
l_n = [i for i in range(10)]
```

```python
l_fn = [factorial(i) for i in range(10)]
```

```python
print(l_n)
```

```python
print(l_fn)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(l_n, l_fn, label='factorial')
plt.xlabel('n')
plt.ylabel('factorial(n)')
plt.legend()
```

```python
def randwalk(n_steps, start=0):
    a_steps=np.random.randint(0,2,size=n_steps)
    a_steps=a_steps*2-1
    pos=np.ndarray(shape=(n_steps,1), dtype=int)
    pos[0]=start
    for i in range(1,n_steps):
        pos[i]=pos[i-1]+a_steps[i]
    plt.plot(range(1, n_steps+1),pos)
```

```python
randwalk(100,0)
```

```python
def random_walk(num_steps,x_start=0):
    a_steps=np.random.randint(2, size=num_steps)
    a_steps=a_steps*2-1
    pos=np.ndarray(shape=(num_steps,1),dtype=int)
    pos[0]=x_start
    for i in range(num_steps):
        pos[i]=pos[i-1]+a_steps[i]
    plt.plot(range(1,num_steps+1), pos)
    plt.xlabel('Number of steps taken')
    plt.ylabel('Position from the starting point')
    plt.legend()
```

```python
random_walk(100,0)
```
