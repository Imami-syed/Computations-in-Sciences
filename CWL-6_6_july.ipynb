## Logistic Map
### Mathematical function Map

```python
import numpy as np
```

```python
def map(x,alpha):
    '''Returns the value of the logistic map y = alpha*x*(1-x)'''
    return alpha*x*(1-x)
```

```python
alpha = 2.0
x0 = 0.1
l1=[x0]
for i in range(2,20):
    l1.append(map(l1[-1],alpha))
print(l1)
```

```python
def logmapseq(x0,alpha,n_iters):
    '''Returns sequence of length n_iters for map with
     parameter value alpha and starting value x0 '''
    l1=[x0]
    for i in range(2, n_iters):
        l1.append(map(l1[-1], alpha))
        
    return l1
```

```python
logmapseq(0.75,2,20)
```

```python
import matplotlib.pyplot as plt
```

```python
x0=0.75
n_iters=100
for a in np.linspace(0,4,8):
    plt.plot(logmapseq(x0,a,n_iters), label='a='+str(a))
    plt.legend()
```

```python
plt.figure(figsize=(18,6))
x0=0.75
n_iters=100
i = 0
for a in np.linspace(0,4,8):
    i+=1
    plt.subplot(2,4,i)
    plt.plot(logmapseq(x0,a,n_iters),'.',label='a=' +str(a))
    plt.legend()
```
