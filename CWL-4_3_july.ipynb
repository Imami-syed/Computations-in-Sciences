```python
import numpy as np
import matplotlib.pyplot as plt
def randwalk(num_steps,x_start=0):
    steps=2*np.random.randint(2,size=num_steps)-1
    return x_start+np.cumsum(steps)
```

```python
rw1=randwalk(10)
print(rw1)
```

```python
n_walks=10
num_steps=100
l_walks=[]
for i in range(n_walks):
    l_walks.append(randwalk(num_steps))
    plt.plot(l_walks[-1])
```

```python
# <x(t)> vs t,  <x^2(t)> vs t, <(dx)^2>
a_walks=np.array(l_walks)
a_walks.shape
```
