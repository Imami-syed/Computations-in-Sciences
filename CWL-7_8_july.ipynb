```python
import numpy as np
def map(x,alpha):
    '''Returns the value of the logistic map y = alpha*x*(1-x)'''
    return alpha*x*(1-x)
```

```python
alpha = 2.0
x0 = 0.1
l1=[x0]
for i in range(2,20):
    l1.append(map(l1[-1],alpha))
print(l1)
```

```python
def logmapseq(x0,alpha,n_iters):
    '''Returns sequence of length n_iters for map with
     parameter value alpha and starting value x0 '''
    l1=[x0]
    for i in range(2, n_iters):
        l1.append(map(l1[-1], alpha))
        
    return l1
```

```python
import matplotlib.pyplot as plt
x0=0.75
n_iters=100
for a in np.linspace(0,4,8):
    plt.plot(logmapseq(x0,a,n_iters), label='a='+str(a))
    plt.legend()
```

```python
xv=np.linspace(0,1,101)
xv1=np.linspace(0,1,101)
```

```python
fv=map(xv,0.5)
fv1=map(xv,0.9)
```

```python
plt.plot(xv,fv)
plt.plot(xv1,fv1)
plt.legend()
```
